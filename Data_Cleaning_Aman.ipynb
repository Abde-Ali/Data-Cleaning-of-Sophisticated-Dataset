```python
for i in df.columns:
  if ((df[i].isnull().sum())/len(df[i]))*100 > 40:
    df.drop(columns=[i],inplace =  True)
  elif (df[i].nunique())==1:
    df.drop(columns=[i],inplace =  True)
  elif (df[i].nunique())==2 :
    if df[i].value_counts()[0]/(df[i].value_counts()[1]+df[i].value_counts()[0])*100 > 70:

      df.drop(columns=[i],inplace =  True)
```
